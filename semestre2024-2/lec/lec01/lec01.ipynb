A menudo, el primer paso en la ciencia de datos es obtener los datos. El siguiente es un pequeño programa para descargar texto de la web.

```python
# Un pequeno programa para descar el texto
def read_url(url): 
    from urllib.request import urlopen 
    import re
    return re.sub('\\s+', ' ', urlopen(url).read().decode())

huck_finn_url = 'https://www.inferentialthinking.com/data/huck_finn.txt'
huck_finn_text = read_url(huck_finn_url)
huck_finn_chapters = huck_finn_text.split('CHAPTER ')[44:]
```

Podemos convertir los resultados de nuestro análisis en más tablas.

```python
counts = Table().with_columns([
    'Tom', np.char.count(huck_finn_chapters, 'Tom'),
    'Jim', np.char.count(huck_finn_chapters, 'Jim'),
    'Huck', np.char.count(huck_finn_chapters, 'Huck'),
])
counts
```

# Aprenderemos a visualizar datos

```python
counts
```

Grafique los recuentos acumulados:

cuántas veces en el capítulo 1, cuántas veces en los capítulos 1 y 2, y así sucesivamente.

```python
df = counts.to_df()
```

```python
# Convert Table to pandas DataFrame for easier plotting
df = counts.to_df()

# Plotting
%matplotlib inline
plt.style.use('fivethirtyeight')
plt.plot(df["Tom"].cumsum(), label="Tom")
plt.plot(df["Jim"].cumsum(), label="Jim")
plt.plot(df["Huck"].cumsum(), label="Huck")
plt.xlabel('Capitulos')
plt.legend()
plt.ylabel('Suma cumulativa')
plt.title('Suma cumulativa de Tom, Jim Huck en los capitulos')
plt.show()
```

We can explore the characters in Little Women using the same kind of analysis.

```python
# Counts of names in the chapters of Little Women
names = ['Amy', 'Beth', 'Jo', 'Laurie', 'Meg']
mentions = {name: np.char.count(little_women_chapters, name) for name in names}
counts = Table().with_columns([
        'Amy', mentions['Amy'],
        'Beth', mentions['Beth'],
        'Jo', mentions['Jo'],
        'Laurie', mentions['Laurie'],
        'Meg', mentions['Meg']
    ])
```

```python
cum_counts = counts.with_column('Chapter', np.arange(1, 48, 1)).to_df()
cum_counts.head(5)
```

```python
Amy = cum_counts["Amy"].cumsum()
Beth = cum_counts["Beth"].cumsum()
Jo = cum_counts["Jo"].cumsum()
Laurie = cum_counts["Laurie"].cumsum()
Meg = cum_counts["Meg"].cumsum()
```

```python
# Plot the cumulative counts
Table.static_plots()
plt.plot(Amy)
plt.plot(Beth)
plt.plot(Jo)
plt.plot(Meg)
plt.title('Cumulative Number of Times Name Appears');
```

# Examining Length

How long are the books? How long are sentences?


```python
len(read_url(huck_finn_url))
```

```python
# In each chapter, count the number of all characters;
# call this the "length" of the chapter.
# Also count the number of periods.

length_hf = Table().with_columns([
        'Length', [len(s) for s in huck_finn_chapters],
        'Periods', np.char.count(huck_finn_chapters, '.')
    ])
length_lw = Table().with_columns([
        'Length', [len(s) for s in little_women_chapters],
        'Periods', np.char.count(little_women_chapters, '.')
    ])
```

```python
# The counts for Huckleberry Finn
length_hf
```

```python
# The counts for Little Women
length_lw
```

```python
Table.static_plots()
plt.figure(figsize=(10,10))
plt.scatter(length_hf[1], length_hf[0], color='darkblue')
plt.scatter(length_lw[1], length_lw[0], color='gold')
plt.xlabel('Number of periods in chapter')
plt.ylabel('Number of characters in chapter');
```
