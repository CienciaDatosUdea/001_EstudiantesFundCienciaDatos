A menudo, el primer paso en la ciencia de datos es obtener los datos. La  siguiente celdas son  un pequeño programa para descargar texto de la web.

````python
import requests
import numpy as np
import matplotlib.pylab as plt
from datascience import *
````

````python
# URL del archivo de texto
url = "https://www.gutenberg.org/cache/epub/2000/pg2000.txt"
# Descargar el contenido del archivo de texto
response = requests.get(url)
text = response.text
# https://www.gutenberg.org/cache/epub/2000/pg2000-images.html
# https://www.gutenberg.org/browse/scores/top-es.php
````

 Capitulos del libro, el quijote de la mancha

````python
quijote_chapter_one = text.split('Capítulo')[1]
quijote_chapters = text.split('Capítulo')[1:]
# Realizar un split del libro
Table().with_column('Chapters', quijote_chapters)
````

````python
# Conteo de la palabra sancho por capítulo

np.char.count(quijote_chapters, 'Sancho')
````

````python
names = ["Sancho","Dulcinea", "Rocinante", "Quijote"]
personajes = Table().with_columns([\
    "Sancho", np.char.count(quijote_chapters, 'Sancho'),
    "Dulcinea", np.char.count(quijote_chapters, 'Dulcinea'),
    "Rocinante",np.char.count(quijote_chapters, 'Rocinante'),
    "Quijote",np.char.count(quijote_chapters, 'Quijote')])
personajes = personajes.to_df()
personajes.iloc[35:47]
````

Analicemos cuantes veces aparece los nombres de los personajes en el libro, para ello apliquemos el concepto de suma cumulativa:

Grafique los recuentos acumulados:

cuántas veces en el capítulo 1, cuántas veces en los capítulos 1 y 2, y así sucesivamente.

```python
# Suma cumulativa
i = np.random.randint(0, 10, 10)
print(i)
print(i.cumsum())
```
--

```python
[4 8 5 7 3 4 8 7 0 1]
[ 4 12 17 24 27 31 39 46 46 47]
```

````python
personajes_cumsum = personajes[["Sancho","Dulcinea", "Rocinante", "Quijote"]].cumsum()
personajes_cumsum
````

````python
%matplotlib inline

plt.style.use('fivethirtyeight')
ax = personajes_cumsum.plot()
plt.xlabel('Capítulos')
plt.ylabel('Suma cumulativa')
plt.title('Suma cumulativa de Sancho, Dulciena, Rocinante, Quijote')
plt.show()
````

Determinemos ahora la relacion entre el número de puntos y la longitud de cada capítulo: 

recordemos que estamos trabajando con la variable ```quijote_chapters```

````python
puntos = np.char.count(quijote_chapters, '.')
longitud = [len(j) for j in quijote_chapters]
capitulos = [j for j in range(0, len(quijote_chapters))]
# Construyamos la tabla 
longitud_puntos = Table().with_columns("puntos",puntos, "longitud", longitud, "capitulos", capitulos)
````

````python
longitud_puntos
````

````python
#plots.figure(figsize=(6, 6))
plt.style.use('fivethirtyeight')
ax=plt.scatter(longitud_puntos.column(0), 
              longitud_puntos.column(1), 
              color='salmon')
plt.xlabel("Número de puntos")
plt.ylabel("Número de caracteres")
````

# Libro de huck_finn_text Mark Twain

````python
# Un pequeno programa para descar el texto
def read_url(url): 
    from urllib.request import urlopen 
    import re
    return re.sub('\\s+', ' ', urlopen(url).read().decode())

huck_finn_url = 'https://www.inferentialthinking.com/data/huck_finn.txt'
huck_finn_text = read_url(huck_finn_url)
huck_finn_chapters = huck_finn_text.split('CHAPTER ')[44:]
````

````python
np.char.count(huck_finn_chapters, 'Tom')
````

````python
counts = Table().with_columns([
    'Tom', np.char.count(huck_finn_chapters, 'Tom'),
    'Jim', np.char.count(huck_finn_chapters, 'Jim'),
    'Huck', np.char.count(huck_finn_chapters, 'Huck'),
])
````

````python
# Convert Table to pandas DataFrame for easier plotting
df = counts.to_df()

# Plotting
%matplotlib inline
plt.style.use('fivethirtyeight')
plt.plot(df["Tom"].cumsum(), label="Tom")
plt.plot(df["Jim"].cumsum(), label="Jim")
plt.plot(df["Huck"].cumsum(), label="Huck")
plt.xlabel('Capitulos')
plt.legend()
plt.ylabel('Suma cumulativa')
plt.title('Suma cumulativa de Tom, Jim, Huck en los capítulos')
plt.show()
````

````python
puntos_huck = np.char.count(huck_finn_chapters, '.')
longitud_huck = [len(j) for j in huck_finn_chapters]
capitulos_huck = [j for j in range(0, len(huck_finn_chapters))]
# Construyamos la tabla 
longitud_puntos_huck = Table().with_columns("puntos",puntos_huck,\
                                            "longitud", longitud_huck, "capitulos", capitulos_huck)
````

````python
#plots.figure(figsize=(6, 6))
plt.style.use('fivethirtyeight')
ax=plt.scatter(longitud_puntos_huck.column(0), 
              longitud_puntos_huck.column(1), 
              color='blue')

plt.xlabel("Número de puntos")
plt.ylabel("Número de caracteres")
````
