**Pregunta 2.7.** 
A partir de una tabla llamada `bakers`, que contiene:  
- Una columna con el número de veces que un concursante ganó *Star Baker* (`star_col`),  
- Una columna que indica si ganó la temporada (`won_col`),  

**¿Cómo escribirías una función que calcule el estadística de prueba adecuado para evaluar si existe una relación entre ganar *Star Baker* y ganar la temporada?**
*Pista:* Asegúrese de tener en cuenta la direccionalidad de nuestra hipótesis alternativa.

````python
# BEGIN SOLUTION NO PROMPT
def prueba_estadistica(bakers, star_col, won_col):  
    media = bakers.group(star_col, np.mean)
    media1 = media.column(won_col)[1]
    media2 = media.column(won_col)[0]
    
    diferencia_observada = media2 - media1
    
    estadistica_prueba = abs(diferencia_observada) 
    return estadistica_prueba

# END SOLUTION
""" # BEGIN PROMPT
def prueba_estadistica(panaderos, label_col, val_col):
    ...
    return estadistica_prueba

""" # END PROMPT
prueba_estadistica(bakers,"won", "star baker awards mean")
````

Si la hipótesis nula es cierta y las distribuciones de premios del panadero estrella son las mismas, esperamos que la diferencia en la media de los premios del panadero estrella no cambie cuando se cambian las etiquetas en "won".

**Pregunta 2.8.** Escriba una función `simulate_and_test_statistic` para calcular una prueba de nuestra prueba A/B. Su función debería ejecutar una simulación y devolver una estadística de prueba.

````python
# BEGIN SOLUTION NO PROMPT

def simulate_and_test_statistic(table, label_column):
    
    shuffled_labels = table.sample(with_replacement=False).column(label_column)
    simulated_table = table.with_column('Shuffled Labels', shuffled_labels).drop(label_column)
    group_means = simulated_table.group('Shuffled Labels', np.mean)[1]
    return abs(group_means[0] - group_means[1])

# END SOLUTION
""" # BEGIN PROMPT
def simulate_and_test_statistic(table, label_column):
    np.random.seed(1)# No borres esta linea
    ...
    ...
    ...
""" # END PROMPT
simulate_and_test_statistic(bakers, "won")
````

**Pregunta 2.10.** Encuentre el valor p para su prueba y asígnelo a `empirical_p`.

````python
# BEGIN SOLUTION NO PROMPT

# Bloque de código para calcular el valor p utilizando la librería datascience

# Vamos a asumir que estamos realizando una prueba estadística 
# y necesitamos un valor p, aunque no usemos los datos reales.
# Aquí, simularemos una comparación entre dos grupos para calcular este p-valor.


# Calculamos el valor p como la proporción de veces que la diferencia simulada
# es mayor o igual a la diferencia observada absoluta
empirical_p = np.count_nonzero(np.abs(differences) >= diferencia_observada) / len(differences)
empirical_p
# END SOLUTION
""" # BEGIN PROMPT
empirical_p = ...
""" # END PROMPT
empirical_p
````

````python
# Supongamos que 'empirical_p' es el valor que debe comprobarse después de realizar el cálculo correspondiente
#empirical_p = 0.05  # valor p esperado después de la prueba
# Asumimos que el estudiante hará el cálculo y asignará el valor a 'empirical_p'
assert isinstance(empirical_p, float), "empirical_p debe ser un número decimal"
assert 0 <= empirical_p <= 0.05, "empirical_p debe estar entre 0 y 1"
````
