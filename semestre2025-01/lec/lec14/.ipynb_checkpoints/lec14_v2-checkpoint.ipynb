# Muestra determinista


Una muestra determinista es un tipo de muestra en la que los elementos se seleccionan de manera no aleatoria, siguiendo criterios o reglas preestablecidas. En otras palabras, no se utiliza el azar para elegir los elementos de la muestra, sino que se aplican ciertos criterios de selección que pueden basarse en características específicas

````python
top.take(3, 8, 19)
````

````python
top.where('Title', are.containing('Harry Potter'))
````

# Muestras de probabilidad


Una muestra de probabilidad es aquella para la que es posible calcular, antes de que se extraiga la muestra, la probabilidad con la que cualquier subconjunto de elementos entrará en la muestra.


Dada una poblacion de 100000, individuos, se puede elegir una muestra de 1000 individuos en este caso, cada individuo tiene una probablidad de ser elegido de 1/10

````python
index = np.random.choice(np.arange(0, top.num_rows), 1000)
top.take(index)
````

# Muestra sistemática: 

Una muestra sistemática es un tipo de muestra de probabilidad en la que los elementos se seleccionan de manera uniforme y regular a lo largo de toda la población. Este método es simple y eficiente cuando se tiene una lista ordenada de la población.


````python
start = np.random.choice(np.arange(10))
index = np.arange(start, top.num_rows, 10)
top.take(index)
````

# Distribuciones empíricas


Una distribución empírica es una representación de los datos observados en una muestra. Se utiliza para describir cómo los valores de los datos están distribuidos sin hacer suposiciones sobre la forma de la distribución subyacente en la población. La distribución empírica se basa directamente en los datos recolectados y permite visualizar y analizar las características clave de la muestra, como su forma, tendencia central y dispersión.

### Un dado

Distribucion teórica para cada cara

# Parámetros y Estadístico

Los datos de población son un todo y completo. La muestra es un subconjunto de la población que se obtiene utilizando el muestreo

Mientras que un parámetro es una función de los datos de la población, el estadístico lo es de los datos de una muestra.





**Promedio :**     Es la media aritmética y se calcula sumando un grupo de números y dividiendo a continuación por el recuento de dichos números. Por ejemplo, el promedio de 2, 3, 3, 5, 7 y 10 es 30 dividido por 6, que es 5.

**Mediana :**    Es el número intermedio de un grupo de números; es decir, la mitad de los números son superiores a la mediana y la mitad de los números tienen valores menores que la mediana. Por ejemplo, la mediana de 2, 3, 3, 5, 7 y 10 es 4.

**Moda:**     Es el número que aparece más frecuentemente en un grupo de números. Por ejemplo, la moda de 2, 3, 3, 5, 7 y 10 es 3.

Supongamos que tienes la siguiente muestra de datos que representa las edades de un pequeño grupo de personas: 

$ \text{Edades: } 22, 25, 22, 30, 28, 22, 25 $

### 1. Media

La **media** es el promedio de los datos y se calcula sumando todos los valores y dividiendo por el número de valores.

¿Cuál es la media?

### 2. Mediana

La **mediana** es el valor que se encuentra en el centro de un conjunto de datos ordenados. Si el número de observaciones es impar, es el valor central. Si es par, es el promedio de los dos valores centrales.

- Primero, ordena los datos: 
  $ 22, 22, 22, 25, 25, 28, 30 $

¿Cuál es la mediana?
### 3. Moda

La **moda** es el valor que aparece con más frecuencia en el conjunto de datos.

- En el conjunto de datos: 
$ 22, 25, 22, 30, 28, 22, 25 $
  
¿Cuál es la moda?

# Estadístico

Es cualquier número calculado a partir de los datos de una muestra. Por lo tanto, la mediana de la muestra es un estadístico.


````python
muestra = united.select("Delay").sample(1000)
np.median(muestra.column("Delay"))
````

````python
np.mean(muestra.column("Delay"))
````

## Simulando un estadístico


1. Elegir un tamaño de la muestra.
2. Calcular el valor del estadístico con la muestra escogida.

````python
def random_sample_median(size):
    return np.median(united.sample(size).column('Delay'))
````

````python
random_sample_median(100)
````

````python
sample_medians = make_array()

for i in np.arange(1000):
    new_median = random_sample_median(10)
    sample_medians = np.append(sample_medians, new_median)
````

````python
medians = Table().with_columns(
    "Sample Medians", sample_medians,
    "Sample Size", 10)
medians.hist("Sample Medians", bins = 50)
````

````python
np.mean(medians.column("Sample Medians"))
````

````python
sample_medians2 = make_array()

for i in np.arange(1000):
    new_median = random_sample_median(1000)
    sample_medians2 = np.append(sample_medians2, new_median)
````

````python
medians.append(Table().with_columns(
    "Sample Medians", sample_medians2,
    "Sample Size", 1000)).hist("Sample Medians", group="Sample Size", bins=50)
plots.xlim(0, 20)
````

simular procesos aleatorios repetidamente es una forma de aproximar las distribuciones de probabilidad sin averiguar las probabilidades matemáticamente o generar todas las posibles muestras aleatorias. Por lo tanto, las simulaciones informáticas se convierten en una poderosa herramienta en la ciencia de datos. Pueden ayudar a los científicos de datos a comprender las propiedades de cantidades aleatorias que serían complicadas de analizar de otras maneras.

Ejemplo:


Supongamos que una escuela tiene un total de 500 estudiantes. Deseas realizar un estudio para determinar el promedio de horas que los estudiantes dedican al estudio por semana.


**1. Definamos la Población y los parámetro de Interés:**

- **Población:** Todos los 500 estudiantes de la escuela.
- **Parámetro de Interés:** La media de horas que todos los estudiantes estudian por semana. Este es el promedio real que querríamos conocer, simbolizado como $ \mu $

**2. Realizar el Muestreo Aleatorio:**

Decides que una muestra de 50 estudiantes será suficiente para tu estudio. Vas a realizar un muestreo aleatorio simple para seleccionar estos estudiantes.

- **Proceso de Selección:**
  - Asigna un número único a cada uno de los 500 estudiantes (del 1 al 500).
  - Utiliza un generador de números aleatorios (o herramientas estadísticas) para elegir 50 números distintos entre 1 y 500.
  - Los estudiantes correspondientes a estos números formarán tu muestra.

**3. Recolectar Datos Muestrales y Calcular Estadísticos:**

Suponiendo que has recolectado la siguiente información de los 50 estudiantes seleccionados, que refleja las horas de estudio por semana:
```python
make_array(15, 10, 12, 14, 16, 17, 13, 11, 18, 12, 15, 14, 16, 14, 13, 17, 15, 16, 12, 13, 14, 15, 12, 11, 17, 12, 15, 14, 14, 11, 13, 17, 14, 16, 15, 11, 13, 12, 14, 18, 15, 13, 12, 14, 15, 16, 17, 12, 15, 13)
```
**4. Calcular la Media Muestral (Estadístico):**

- **Media Muestral $(\bar{x}$ :** Es el promedio de horas de estudio de los 50 estudiantes en la muestra.



- La **media muestral** $ \bar{x} $ horas es un **estadístico**, que estima el parámetro poblacional \( \mu \), la media real de las horas de estudio por semana para todos los estudiantes de la escuela.
