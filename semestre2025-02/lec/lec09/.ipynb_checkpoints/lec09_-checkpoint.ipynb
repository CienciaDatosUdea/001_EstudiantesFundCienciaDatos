# Datos de encicla 

````python
en_cicla = Table.read_table("Datos_de_Estaciones_Mejorados.csv")
en_cicla
````

````python
en_cicla_df = en_cicla1.to_df()
````

````python
en_cicla_df
````

Realicemos un calculo teórico de los tiempos entre estaciones

````python
# Si en_cicla es un objeto de tipo Table, lo convertimos a DataFrame
en_cicla_df = en_cicla.to_df()

# Asegurarse de que las columnas de Calle y Carrera sean numéricas
en_cicla_df['Calle'] = pd.to_numeric(en_cicla_df['Calle'], errors='coerce')
en_cicla_df['Carrera'] = pd.to_numeric(en_cicla_df['Carrera'], errors='coerce')

# Definir la velocidad en km/h
velocidad_kmh = 25  # 15 km/h

# Función para calcular la distancia aproximada entre dos estaciones
def calcular_distancia(calleA, carreraA, calleB, carreraB):
    return (abs(calleA - calleB) + abs(carreraA - carreraB))*0.08

# Crear una lista para almacenar los tiempos
tiempos_estimados = []

# Generar las combinaciones entre todas las estaciones
for i, estacionA in en_cicla_df.iterrows():
    for j, estacionB in en_cicla_df.iterrows():
        if i != j:
            distancia = calcular_distancia(estacionA['Calle'], estacionA['Carrera'], estacionB['Calle'], estacionB['Carrera'])
            # Convertir la distancia en kilómetros y calcular el tiempo en horas
            tiempo = (distancia / velocidad_kmh) * 60  # Tiempo en minutos
            tiempos_estimados.append({
                'startA': estacionA['NOMBRE ESTACION'],
                'startB': estacionB['NOMBRE ESTACION'],
                'Distancia': distancia,
                'Tiempo (min)': tiempo
            })

# Convertir los tiempos en un DataFrame
df_tiempos = pd.DataFrame(tiempos_estimados)

# Mostrar los resultados
df_tiempos.head()
````

````python
df_tiempos_clean = df_tiempos.dropna()

# Mostrar los resultados sin NaN
df_tiempos_clean.head()
````

````python
df_tiempos_clean.to_csv("tiempos_de_viaje_teoricos.csv")
````
