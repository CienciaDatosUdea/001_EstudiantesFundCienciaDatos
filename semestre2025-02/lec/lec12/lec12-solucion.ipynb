# Problema The monty hall 




El problema de Monty Hall es un clsico de la teora de probabilidades y es famoso por ilustrar cmo la intuicin humana a menudo puede llevar a conclusiones incorrectas acerca de las probabilidades. Est basado en un programa de concursos televisivo llamado "Lets Make a Deal", que fue presentado por Monty Hall. A continuacin se presenta una introduccin gua al problema:

### Descripcin del Problema:

Imagina que participas en un concurso donde se te presenta una eleccin entre tres puertas. Detrs de una de estas puertas hay un premio muy deseado, como un coche, y detrs de las otras dos puertas hay un premio indeseado, como una cabra.

**Pasos del Concurso:**
1. **Eleccin Inicial:** Primero, se te pide que elijas una de las tres puertas.
2. **Apertura del Presentador:** Luego de que elijas una puerta, el anfitrin (quien sabe lo que hay detrs de todas las puertas) abrir una de las otras dos puertas, revelando una cabra.
3. **Segunda Eleccin:** Ahora tienes la opcin de quedarte con tu eleccin inicial o cambiar a la otra puerta restante.

### La Cuestin:

La pregunta fundamental que plantea el problema es: Deberas cambiar tu eleccin o quedarte con tu eleccin inicial para maximizar tus posibilidades de ganar el coche?

# Algoritmo 

1. Definir el estado inicial, como van estar las puertas, con las cabras y el carro.
2. Esta eleccin puede estar de forma aleatoria
3. Tomar la eleccion del concursante.
4. Si eleccion del concursante == Cabra 1
    eleccion del concursante, Cabra 2, Carro
4. Si no, si: eleccion del concursante == Cabra 2
    eleccion del concursante, Cabra 1, Carro
5. Si no, si :eleccion del concursante == Carro
   carro, cabra 1, cabra 2


Lo anterior puede ser definido dentro de una funcin y repetido N-veces


```python
from datascience import *
import numpy as np
```

```python
def monty_hall():
    puertas = make_array("Cabra Alice", "Cabra Bob", "Carro")
    np.random.shuffle(puertas)
    eleccion_jugador = np.random.choice(puertas)
    
    if eleccion_jugador=="Cabra Alice":
        revelada = "Cabra Bob"
        OtraOpcion= "Carro"
    
    elif eleccion_jugador=="Cabra Bob":
        revelada = "Cabra Alice"
        OtraOpcion= "Carro"
    
    elif eleccion_jugador=="Carro":
        e = np.random.choice(make_array("Cabra Alice", "Cabra Bob"))
        if(e=="Cabra Alice"):
            revelada = e
            OtraOpcion="Cabra Bob"
        elif(e=="Cabra Bob"):
            revelada = e
            OtraOpcion= "Cabra Alice"
    return make_array(eleccion_jugador,revelada, OtraOpcion) 
games = Table(['Adivinar', 'Revelada', 'OtraOpcion'])
for i in range(0, 10000):
    resultados = monty_hall()
    games.append(resultados)
```
